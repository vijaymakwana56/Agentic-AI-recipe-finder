```python
def contains_hindi(text):
    for char in text:
        if '\u0900' <= char <= '\u097F':
            return True
    return False
```

```python
import requests

# 1. Replace with your actual token from Hugging Face
# MODEL_ID = "BAAI/bge-small-en-v1.5"
MODEL_ID = "BAAI/bge-m3"
# API_URL = f"https://router.huggingface.co/hf-inference/models/{MODEL_ID}"
API_URL= f"https://router.huggingface.co/hf-inference/models/{MODEL_ID}/pipeline/feature-extraction"

def get_embeddings_batch(text_list):
    """Sends a list of texts to the API and returns a list of vectors."""
    headers = {
        "Authorization": f"Bearer {HF_TOKEN}",
        "Content-Type": "application/json",
        "X-Wait-For-Model": "true" # Forces the API to load the model if it's idle
    }
    payload = {
        "inputs": text_list,
        # "options": {"wait_for_model": True}
    }
    response = requests.post(API_URL, headers=headers, json=payload)

    if response.status_code == 200:
        return response.json()
    else:
        print(f"Error {response.status_code}: {response.text}")
        return None
```

```python
import time

# --- BATCH EXECUTION ---
batch_size = 10
embedding_vectors = []

# Assuming your combined text is in a column named 'text'
text_data = df['text'].tolist()

print(f"Starting batch embedding for {len(text_data)} recipes...")

for i in range(0, len(text_data), batch_size):
    batch = text_data[i : i + batch_size]
    vectors = get_embeddings_batch(batch)

    if vectors:
        embedding_vectors.extend(vectors)
    else:
        # If a batch fails, we add None to keep the list length consistent
        embedding_vectors.extend([None] * len(batch))

    # Optional: Brief pause to be nice to the API
    time.sleep(0.1)

# Add the vectors back to your original DataFrame
# df['embedding'] = embedding_vectors
```

```python
from qdrant_client import QdrantClient
from qdrant_client.models import Distance, VectorParams, PointStruct
import uuid
```

```python
#initializing the client to qdrant
client = QdrantClient(
    url=QD_URL,
    api_key=QD_API_KEY
)
```

```python
client.recreate_collection(
    collection_name="indian_recipes",
    vectors_config = VectorParams(size=1024, distance=Distance.COSINE)
)
```

```python
#converting the text and vector pairs into points to add to qdrant
point = []
for text_val, vector_val in zip(text_data, embedding_vectors):
  if vector_val is not None:
    point.append(
        PointStruct(
            id=str(uuid.uuid4()), #generating unique id as strings
            vector=vector_val,
            payload={
                "recipe": text_val #storing the full recipe
            }
        )
    )
```

```python
len(point)
```

```python
# 3. Push to Qdrant in a loop in batches of 100
batch_size = 200
for i in range(0,len(point),batch_size):
  batch = point[i : i + batch_size]

  client.upsert(
      collection_name="indian_recipes",
      points=batch
  )
```
